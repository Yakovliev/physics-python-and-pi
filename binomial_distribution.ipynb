```python
import numpy as np
import plotly.graph_objects as go
```

```python
# Example 1: Flipping a Coin
n_trials_coin = 2
p_heads = 0.6
n_simulations = 10000
coin_flips = np.random.binomial(n_trials_coin, p_heads, n_simulations)
```

```python
# Example 2: Quality Control in Manufacturing
n_items = 100
p_defective = 0.01
n_simulations_qc = 10000
defects = np.random.binomial(n_items, p_defective, n_simulations_qc)
# defects = defects / n_simulations_qc
```

```python
# Plotting the results using Plotly for the Coin Flips

# histnorm='percent' normalizes the histogram to show percentages instead of counts
fig1 = go.Figure()
fig1.add_trace(go.Histogram(x=coin_flips, nbinsx=(n_trials_coin+1), name='Coin Flips', marker_color='blue', histnorm='percent'))
fig1.update_layout(
    title=f"Coin Flipping Simulation. Coin Flipped {n_trials_coin} times. Probability of Heads = {p_heads}. Number of Simulations = {n_simulations}.",
    xaxis_title="Number of Heads",
    yaxis_title="Frequency",
    bargap=0.2
)
fig1.show()
```

```python
# Plotting the results using Plotly for the Quality Control in Manufacturing
fig2 = go.Figure()
fig2.add_trace(go.Histogram(x=defects, nbinsx=(n_items+1), name='Defects in Manufacturing', marker_color='red'))
fig2.update_layout(
    title=f"Manufacturing Defects Simulation. {n_items} Items Produced. Probability of Defective Item = {p_defective}. Number of Simulations = {n_simulations_qc}.",
    xaxis_title="Number of Defective Items",
    yaxis_title="Frequency",
    bargap=0.1
)
fig2.show()
```
