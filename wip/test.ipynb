```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Constants
k_B = 1.38e-23  # Boltzmann constant (J/K)
T = 298  # Temperature (K)
gamma = 1.0e-12  # Friction coefficient (NÂ·s/m)
D = k_B * T / gamma  # Diffusion coefficient (m^2/s)
dt = 1e-15  # Time step (s)
n_steps = 100000  # Number of steps
```

```python
# Initialize position
position = np.zeros((n_steps, 3))

# Brownian Dynamics loop
for t in range(1, n_steps):
    random_force = np.random.normal(0, np.sqrt(2 * D * dt), 3)
    position[t] = position[t-1] + random_force

# Calculate Mean Squared Displacement (MSD)
msd = np.sum((position - position[0])**2, axis=1)

# Time array
time = np.arange(n_steps) * dt
```

```python
# Plot MSD vs. time
plt.plot(time, msd)
plt.xlabel('Time (s)')
plt.ylabel('Mean Squared Displacement (m^2)')
plt.title('MSD of CNâ» ion in water')
plt.show()

# Estimate diffusion coefficient
D_estimated = np.polyfit(time, msd, 1)[0] / 6
print(f"Estimated Diffusion Coefficient: {D_estimated:.2e} m^2/s")
```
